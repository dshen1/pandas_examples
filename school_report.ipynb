```python
import pandas as pd
```

```python
xl=pd.ExcelFile('DRC_Raw_Data2010-11.xls')
```

```python
basic=xl.parse('Basic Data', skiprows=15)
```

```python
basic.groupby('statename').agg({'totschools':sum}).sort_values(['totschools']).stack().unstack('statename')
```

```python
%matplotlib inline
```

```python
basic.groupby('statename').agg({'totschools':sum}).sort_values(['totschools']).sort('totschools', ascending=False).stack().unstack('statename').plot(kind='barh')
```
